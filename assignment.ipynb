```python
def clean_text(text):
    # Specify the text pieces to remove
    pieces_to_remove = [
    'We provide intelligence, accelerate innovation and implement technology with extraordinary breadth and depth global insights into the big data,data-driven dashboards, applications development, and information management for organizations through combining unique, specialist services and high-lvel human expertise.',
    'Contact us: ',
    '© All Right Reserved, Blackcoffer(OPC) Pvt. Ltd'
]

    
    
    # Remove escape characters
    text = re.sub(r'\\[\'"\\]', '', text)
    
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    
    # Remove 'xa0'
    text = text.replace('xa0', ' ')
    
    # Remove each specified text piece using regular expression
    if pieces_to_remove:
        for piece in pieces_to_remove:
            text = re.sub(re.escape(piece), '', text)
    
    # Remove punctuations
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    return text

```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
import numpy as np

# Assuming you have loaded your DataFrame 'df' and defined positive_words, negative_words, and custom_stopwords

# Function to calculate sentiment scores
from nltk.tokenize import word_tokenize

def calculate_scores(text, positive_words, negative_words, custom_stopwords):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove custom stopwords
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in custom_stopwords]

    print("Filtered Tokens:", filtered_tokens)  # Add this line for debugging

    # Calculate Positive Score
    positive_score = sum(1 for word in filtered_tokens if word in positive_words)

    # Calculate Negative Score
    negative_score = sum(1 for word in filtered_tokens if word in negative_words)

    print("Positive Score:", positive_score)  # Add this line for debugging
    print("Negative Score:", negative_score)  # Add this line for debugging

    # Calculate Polarity Score
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)

    # Calculate Subjectivity Score
    subjectivity_score = (positive_score + negative_score) / (len(filtered_tokens) + 0.000001)

    return positive_score, negative_score, polarity_score, subjectivity_score

# Create new columns for each score
df['Positive Score'], df['Negative Score'], df['Polarity Score'], df['Subjectivity Score'] = \
zip(*df['Text'].apply(lambda x: calculate_scores(x, positive_words, negative_words, custom_stopwords)))

# Display the DataFrame with scores
print(df[['Positive Score', 'Negative Score', 'Polarity Score', 'Subjectivity Score']])
```

```python
def calculate_readability(text):
    # Tokenize the text into sentences and words
    sentences = sent_tokenize(text)
    words = word_tokenize(text)

    # Check if there are sentences and words
    if len(sentences) == 0 or len(words) == 0:
        return 0.0, 0.0, 0.0  # Return zeros to avoid division by zero

    # Calculate Average Sentence Length
    avg_sentence_length = len(words) / len(sentences)

    # Define your criteria for complex words (you can customize this)
    complex_words = [word for word in words if len(word) > 6]  # Example: words with more than 6 characters

    # Check if there are words to avoid division by zero
    if len(words) == 0:
        return avg_sentence_length, 0.0, 0.0  # Return zeros for Percentage of Complex Words and Fog Index

    # Calculate Percentage of Complex words
    percentage_complex_words = len(complex_words) / len(words)

    # Calculate Fog Index
    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

    return avg_sentence_length, percentage_complex_words, fog_index

# Create new columns for readability metrics
df['Average Sentence Length'] = 0.0
df['Percentage of Complex Words'] = 0.0
df['Fog Index'] = 0.0

# Iterate over each row and calculate readability metrics
for index, row in df.iterrows():
    avg_sentence_length, percentage_complex_words, fog_index = calculate_readability(row['Text'])
    df.at[index, 'Average Sentence Length'] = avg_sentence_length
    df.at[index, 'Percentage of Complex Words'] = percentage_complex_words
    df.at[index, 'Fog Index'] = fog_index

# Display the DataFrame with readability metrics
print(df[['Average Sentence Length', 'Percentage of Complex Words', 'Fog Index']])
```

```python
# Function to calculate the average number of words per sentence
def calculate_average_words_per_sentence(text):
    # Tokenize the text into sentences and words
    sentences = sent_tokenize(text)
    words = word_tokenize(text)

    # Check if there are sentences to avoid division by zero
    if len(sentences) == 0:
        return 0.0  # Return zero to avoid division by zero

    # Calculate Average Number of Words Per Sentence
    average_words_per_sentence = len(words) / len(sentences)

    return average_words_per_sentence

# Create a new column for the average number of words per sentence
df['Average Words Per Sentence'] = 0.0

# Iterate over each row and calculate the average number of words per sentence
for index, row in df.iterrows():
    avg_words_per_sentence = calculate_average_words_per_sentence(row['Text'])
    df.at[index, 'Average Words Per Sentence'] = avg_words_per_sentence

# Display the DataFrame with the new column
print(df[['Average Words Per Sentence']])
```

```python
# Function to count the number of syllables in a word
def count_syllables(word):
    vowels = "aeiouy"
    count = 0
    last_char_was_vowel = False

    for char in word.lower():
        if char in vowels:
            if not last_char_was_vowel:
                count += 1
            last_char_was_vowel = True
        else:
            last_char_was_vowel = False

    return count

# Function to check if a word is complex (has more than two syllables)
def is_complex(word):
    return count_syllables(word) > 2

# Assuming you have loaded your DataFrame 'df'

# Create a new column for the complex word count
df['Complex Word Count'] = 0

# Iterate over each row and calculate the complex word count
for index, row in df.iterrows():
    # Tokenize the text into words
    words = word_tokenize(row['Text'])
    
    # Count the number of complex words
    complex_word_count = sum(1 for word in words if is_complex(word))
    
    # Update the DataFrame with the complex word count
    df.at[index, 'Complex Word Count'] = complex_word_count

# Display the DataFrame with the new column
print(df[['Complex Word Count']])
```

```python
# Function to calculate Word Count
def calculate_word_count(text, custom_stopwords):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove custom stopwords and punctuations
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in custom_stopwords and word.lower() not in string.punctuation]

    # Calculate Word Count
    word_count = len(filtered_tokens)

    return word_count

# Create a new column for Word Count
df['Word Count'] = 0

# Iterate over each row and calculate the Word Count
for index, row in df.iterrows():
    word_count = calculate_word_count(row['Text'], custom_stopwords)
    
    # Update the DataFrame with the Word Count
    df.at[index, 'Word Count'] = word_count

# Display the DataFrame with the new column
print(df[['Word Count']])
```

```python
# Function to calculate Syllable Count Per Word
def calculate_syllable_count_per_word(text, custom_stopwords):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove custom stopwords and punctuations
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in custom_stopwords and word.lower() not in string.punctuation]

    # Calculate Syllable Count Per Word
    syllable_count_per_word = 0
    for word in filtered_tokens:
        # Count syllables using a simple heuristic (counting vowels)
        vowels = "aeiouy"
        count = 0
        last_char = None
        for char in word:
            char = char.lower()
            if char in vowels and (last_char is None or last_char not in vowels):
                count += 1
            last_char = char
        
        # Handle exceptions for words ending with "es" or "ed"
        if word.endswith("es") or word.endswith("ed"):
            count -= 1

        syllable_count_per_word += max(1, count)

    return syllable_count_per_word

# Create a new column for Syllable Count Per Word
df['Syllable Count Per Word'] = 0

# Iterate over each row and calculate the Syllable Count Per Word
for index, row in df.iterrows():
    syllable_count_per_word = calculate_syllable_count_per_word(row['Text'], custom_stopwords)
    
    # Update the DataFrame with the Syllable Count Per Word
    df.at[index, 'Syllable Count Per Word'] = syllable_count_per_word

# Display the DataFrame with the new column
print(df[['Syllable Count Per Word']])
```

```python
# Function to calculate count of personal pronouns
def calculate_personal_pronouns(text):
    # Define a regular expression pattern for personal pronouns
    pronoun_pattern = re.compile(r'\b(?:I|we|my|ours|us)\b', flags=re.IGNORECASE)

    # Find all matches in the text
    matches = pronoun_pattern.findall(text)

    # Return the count of personal pronouns
    return len(matches)

# Create a new column for Personal Pronouns Count
df['Personal Pronouns Count'] = 0

# Iterate over each row and calculate the count of Personal Pronouns
for index, row in df.iterrows():
    pronoun_count = calculate_personal_pronouns(row['Text'])
    
    # Update the DataFrame with the count of Personal Pronouns
    df.at[index, 'Personal Pronouns Count'] = pronoun_count

# Display the DataFrame with the new column
print(df[['Personal Pronouns Count']])
```

```python
# Function to calculate average word length
def calculate_average_word_length(text):
    # Tokenize the text into words
    words = re.findall(r'\b\w+\b', text)
    
    # Calculate the total number of characters in all words
    total_characters = sum(len(word) for word in words)
    
    # Calculate the total number of words
    total_words = len(words)
    
    # Calculate the average word length
    average_word_length = total_characters / total_words if total_words > 0 else 0
    
    return average_word_length

# Create a new column for Average Word Length
df['Average Word Length'] = 0.0

# Iterate over each row and calculate the Average Word Length
for index, row in df.iterrows():
    avg_word_length = calculate_average_word_length(row['Text'])
    
    # Update the DataFrame with the Average Word Length
    df.at[index, 'Average Word Length'] = avg_word_length

# Display the DataFrame with the new column
print(df[['Average Word Length']])
```
